```python
from selenium import webdriver
from selenium.webdriver.support.wait import WebDriverWait
import time
from time import sleep
from bs4 import BeautifulSoup
import pandas as pd

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver =webdriver.Chrome('chromedriver',chrome_options=chrome_options)
url='https://www.imdb.com/title/tt7286456/reviews?ref_=tt_urv'
wait=WebDriverWait(driver,10)
driver.get(url)

for i in range(19):
  button=driver.find_element_by_class_name('ipl-load-more__button')
  time.sleep(2)
  button.click()
  time.sleep(5)
time.sleep(10)
data=driver.page_source
driver.quit()
soup=BeautifulSoup(data,'html.parser')


def user_name(soup):
  names=[]
  for div in soup.find_all(name='div',attrs={"class":"display-name-date"}):
    for span in div.find_all(name='span',attrs={"class":"display-name-link"}):
      names.append(span.text)
  return names

a=user_name(soup)

def review(soup):
  reviews=[]
  for div in soup.find_all(name='div',attrs={'class':'content'}):
    for di in div.find_all(name='div',attrs={'class':'show-more__control'}):
      reviews.append(di.text)
  return reviews

b=review(soup)

zippedList = list(zip(a,b))
df = pd.DataFrame(zippedList, columns=['User Name','Review'])
df
```

```python
#data cleaning and pre processing
import string
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
words=stopwords.words('english')
from textblob import Word
from textblob import TextBlob
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import PorterStemmer
st=PorterStemmer()
```

```python
df['lower_case']=df['Review'].apply(lambda x:" ".join(x.lower() for x in x.split()))
df['punctuation']=df['lower_case'].apply(lambda  x: " ".join(x for x in x.split() if x not in string.punctuation))
df['special_characters']=df['punctuation'].apply(lambda x:" ".join(x.replace('[#,@,&]', '') for x in x.split()))
df['stop_words']=df['special_characters'].apply(lambda x:" ".join(x for x in x.split() if x not in words))
df['numbers']=df['stop_words'].apply(lambda x:" ".join(x.replace('\d+', '') for x in x.split()))
df['spelling']=df['numbers'].apply(lambda x: str(TextBlob(x).correct()))
df['token']=df['spelling'].apply(lambda x: TextBlob(x).words)
df['stemming']=df['token'].apply(lambda x: " ".join([st.stem(word) for word in x]))
df['lemmatization']=df['spelling'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))

df
```

```python
file=df.to_csv('cleaned_data.csv',header=True,index=None)
                                                 
```

```python
nltk.download('averaged_perceptron_tagger')
#df_1=pd.read_csv('/content/cleaned_data.csv')
df['lemmatize tokens']=df['lemmatization'].apply(lambda x: TextBlob(x).words)
tag_pos=[]
for i in df['lemmatize tokens']:
  tag_pos.append(nltk.pos_tag(i))

tag_pos
```

```python
def total(tag_pos):
  nn=0
  dt=0
  verb=0
  cord_conj=0
  pro=0
  pos_end=0
  adv=0
  adj=0
  prep=0
  pre_det=0

  nouns=[]
  det=[]
  verbs=[]
  coordinate_conjuction=[]
  pronoun=[]
  posend=[]
  adverb=[]
  adjective=[]
  preposition=[]
  predet=[]
  for i in tag_pos:
    for j in i:
      if(j[1]=='NN'):
        nn+=1
      elif (j[1]=='DT'):
        dt+=1
      elif (j[1]=='VB'):
        verb+=1
      elif (j[1]=='CC'):
        cord_conj+=1
      elif (j[1]=='PRP'):
        pro+=1
      elif (j[1]=='POS'):
        pos_end+=1
      elif (j[1]=='RB'):
        adv+=1
      elif (j[1]=='JJ'):
        adj+=1
      elif (j[1]=='IN'):
        prep+=1
      elif (j[1]=='PDT'):
        pre_det+=1
    nouns.append(nn)
    det.append(dt)
    verbs.append(verb)
    coordinate_conjuction.append(cord_conj)
    pronoun.append(pro)
    posend.append(pos_end)
    adverb.append(adv)
    adjective.append(adj)
    preposition.append(prep)
    predet.append(pre_det)
    nn=0
    dt=0
    verb=0
    cord_conj=0
    interjection=0
    pos_end=0
    adv=0
    adj=0
    prep=0
    pre_det=0

  return nouns,det,verbs,coordinate_conjuction,pronoun,posend,adverb,adjective,preposition,predet

t=total(tag_pos)
```

```python
import pandas as pd

zippedlist=list(zip(t[0],t[1],t[2],t[3],t[4],t[5],t[6],t[7],t[8],t[9]))
df_total = pd.DataFrame(zippedlist, columns=['Nouns','Determiner','Verbs','Coordinating conjunction','Personal pronoun','Possessive ending','Adverb','Adjective','Preposition','Predeterminer'])
df_total.head(5)
```
